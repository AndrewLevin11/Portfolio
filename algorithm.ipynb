# Project Description
There are IPs on the allow list that need to be removed. They are in the remove list
And an algorithm needs to be made to remove the said IPs from the allow list. 
The allow list is made of IPs that have access to sensitive information in an organization
In a cybersecurity job a common task might be to remove the IPs of people who are no longer with the organization
Thus a remove list is created consisting of said IPs that are to be removed from the allow list


```python
remove_list = ["192.168.1.1"]

# Open the file that contains the allow list
# Read the file contents
with open("allow_list.txt", "r") as f:
    ip_addresses = f.read()

# Convert the string into a list
ip_addresses = ip_addresses.split()

# Iterate through the remove list
# Remove IP addresses that are on the remove list
for element in remove_list:
    ip_addresses.remove(element)

# Update the file with the revised list of IP addresses
with open("allow_list.txt", "w") as f:
    text = "\n".join(ip_addresses)
    f.write(text)
```

# Summary
Opened the allow list in read mode, to extract the IPs listed within which are in string format. Converted the string into a list via the split function. 
Created a for loop to remove each IP in the remove list that is in the allow list. Converted the list back into a string via the join function.  
Opened the allow list in write mode to overwrite the previous allow list, now with the remove list IPs taken out.
